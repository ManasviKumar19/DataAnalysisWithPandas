### Annotations and Drawing on a subplot

```python
# It is a bit tricky with all the code that is used, but nevertheless you will understand
```

```python
fig,ax = plt.subplots()

daa = pd.read_csv('examples/spx.csv',index_col=0, parse_dates=True)

sphynx = daa['SPX']

sphynx.plot(color = 'k' )




crisis_day = [(datetime(2007,10,11),'peak of bull market'),
              (datetime(2008,3,12),'Bear sterns fall'),
              (datetime(2008,9,15),'The Big Short')
]



for date,label in crisis_day:
    ax.annotate(label,xy = (date,sphynx.asof(date)+70), # This denotes where the (x,y) coordinate will be located
                xytext = (date , sphynx.asof(date)+225), # This denotes where the text will be located
                arrowprops = dict(facecolor = 'cyan',headwidth = 20,width = 10,headlength = 10),
                horizontalalignment = 'left' , verticalalignment = 'center'
               )
    
    
# Now, to just focus on the market crash years i.e. from 2007 to 2010

ax.set_xlim(['1/1/2007','1/1/2010']) # This will automatically convert the data into year and divide it in equal sets starting
# from 2007 - 01 to 2011 - 01

ax.set_ylim([600,2000])
```

### Scatter or Point Plots

```python
# Now, Scatter plots are especially useful for examining the relation between two one dimensional data series
```

```python
# Here we will perform some computations on a macro data and see the projections
```

```python
macro = pd.read_csv('examples/macrodata.csv')

macro
```

```python
dataless = macro[['cpi','m1','tbilrate','unemp']]

dataless
```

```python
computed_data=np.log(dataless).diff().dropna()  # Here diff() is used to calculate nth order differential

computed_data
```

```python
# Now we will use the seaborn's regplot that makes a scatter plot and passes a linear regression line for checking the closeness
# By closeness what i mean is we can find where and which values are most abundant
```

```python
plots = sns.regplot(x = 'm1',y = 'unemp',data=computed_data,color='r')

```

```python
# Now, for much better analysis it is better to look at scatter plots among a group of variables
# So seaborn has a pair plot function that pairs the density estimate of each variable along the diagonal
```

```python
newplot = sns.pairplot(computed_data,diag_kind='kde',plot_kws={'alpha':0.5})

# Now here kde is a type of diagonal kind which I think means is the gaussian model
# The plot_kws argument is used for passing configuration options to individual plotting calls on the off-diagonal elements
# You can check more in the docs
```
