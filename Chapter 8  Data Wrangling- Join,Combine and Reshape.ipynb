## Combining and Merging Datasets

### Database style DataFrame Joins

#### pandas.merge

```python
df = pd.DataFrame({'kiss':['a','b','b','b','a','a','a','c'],
                  'count':pd.Series(range(8))
                   
                  })

df
```

```python
dff = pd.DataFrame({'kiss':['a','b','b','a'],
                  'count1':pd.Series(range(4))
                  })
dff
```

```python
pd.merge(df,dff) # Now if you dont specify the column on which to join, it will by default choose the one that has overlapping values
```

```python
pd.merge(df,dff,on = 'kiss') # Its important to specify
```

```python
fa = pd.DataFrame({'lkiss':['a','b','b','b','a','a','a','c'],
                  'count':pd.Series(range(8))
                   
                  })
fa
```

```python
af = pd.DataFrame({'rkiss':['a','b','b','a'],
                  'count1':pd.Series(range(4))
                  })

af
```

```python
pd.merge(fa,af,left_on = 'lkiss',right_on = 'rkiss') # Now there are no common cols, so you will have to specify and

# the most important thing is the order of the Dataframes that you pass in merge, like here the first is fa, so left_on will be lkiss
```

```python
# By default merge does inner join only
```

```python
pd.merge(df,dff,how = 'outer')
```

```python
pd.merge(fa,af,left_on = 'lkiss',right_on = 'rkiss',how = 'outer')
```

```python
pd.merge(df,dff,how = 'left')
```

##### To do join on multiple columns

```python
left = pd.DataFrame({'key1':['aaj','mai','khush','hu'],
                     'key2':['ek','ek','do','ek'],
                     'lval':pd.Series([1,2,3,4])
                    })

left
```

```python
right = pd.DataFrame({'key1':['aaj','mai','bahaut','zyada','khush','hu'],
                      'key2':['ek','do','teen','ek','do','do'],
                      'rval':pd.Series([6,7,8,9,10,11])
                     })
right
```

```python
pd.merge(left,right,on = ['key1','key2'],how = 'outer')
```

```python
# Now suppose you apply join only on one column

pd.merge(left,right,on = 'key1',how = 'outer')    
```

```python
# So by default it will make the first one as _x and then 2nd one as _y, to put your custom name do this
```

```python
pd.merge(left,right,on = 'key1',how = 'outer',suffixes = ['_pehlla','_dusra'])    
```

### Pivoting "Long" to "Wide" Format

#### It is basically used while storing data where values are stored in one single row rather than multiple values per row

```python
daa = pd.read_csv('macrodata.csv')
```

```python
daa
```

```python
daa = daa.loc[:,['year','quarter','realgdp','infl','unemp']]

daa
```

```python
# Now here we will do some data manipulation and cleaning
```

```python
# First we will start with PeriodIndex, this will convert the year column into time interval
```

```python
# While doing PeriodIndex we will pop the data from the original DataFrame and add it back again after the upgrade
```

```python
periods = pd.PeriodIndex(year = daa.pop('year'),
                        quarter = daa.pop('quarter'),
                        name = 'dates'
                        )

periods
```

```python
daa.index = periods.to_timestamp('D')
```

```python
daa
```

```python
daa.index
```

```python
daa = daa.reindex(columns = ['realgdp','infl','unemp'])
```

```python
daa.columns.name = 'Item Number'
```

```python
daa
```

```python
# Finally we will convert the DF using stack, and then using reindex we will add the row indexes to columns
# and also give the data name as value
```

```python
long = (daa.stack().reset_index().rename(columns={0:'value'}))

long
```

```python
# Now suppose you want to go back to the normal reprsentation done column wise, to do this you use pivot method
```

```python
pivot = long.pivot(index='dates',columns= 'Item Number',values= 'value')

pivot
```

```python
# Now here you mainly have one column with the value. Suppose you had two columns
```

```python
long['dusra value'] = np.random.standard_normal(len(long['value']))

long
```

```python
pivot = long.pivot(index='dates',columns= 'Item Number') # Now here don't pass the value attribute and it will make it into 
# hiearchial DataFrame
 
pivot
```

### Pivoting "Wide" to "Long" Format

#### This is just like the inverse of the one we studied earlier

```python
df = pd.DataFrame({'key':['foo','bar','set'],
                   'A':[1,2,3],
                   'B':[4,5,6],
                   'C':[7,8,9]
                   
                  })

df
```

```python
# Now when using the inverse that is pandas.melt you need to pass an indicator column and other columns will be the values
```

```python
meltme = pd.melt(df,id_vars='key')

meltme
```

```python
# Now here, we will use pivot to get back the original DataFrame
```

```python
og = meltme.pivot(index='key',columns='variable',values='value')
og
```

```python
# Now as you can see that the column has been converted into index so to set it back do
```

```python
og.reset_index()
```

```python
# You can also specify the subset the column values to be used
```

```python
newmelt = pd.melt(df,id_vars='key',value_vars=['A','C'])

newmelt
```

```python
# You can also do it without id_vars
```

```python
melted = pd.melt(df)

melted
```
