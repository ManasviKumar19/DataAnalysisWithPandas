### csv.dialect sub class

```python
class my_dialect(cv.Dialect):
    
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = cv.QUOTE_MINIMAL
    
```

```python
reader = cv.reader(fo, dialect = my_dialect)

for i in reader:
    print('enter')
    print(i)
print('aaka')
```
